```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
#either convert the units [write function and convert]or ignore
#range is taken avg
def convert_sqft_to_num(x):
    token=x.split('-')
    if len(token)==2:
        return (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except:
        return None
  
convert_sqft_to_num('100-200')
```

```python
 convert_sqft_to_num('34.46sq.Meter')#return nothing
```

```python
df4=df3.copy()#deep copy
df4['total_sqft']=df4['total_sqft'].apply(convert_sqft_to_num)
df4.head(10) #other are ignore here like sq.meter
```

```python
#remove those outliers
df6=df5[~(df5.total_sqft/df5.bhk<300)]
df6.shape
```

```python
df6.price_per_sqft.describe()
```

```python
#remove extreme case like min and max value 
#use std deviation value near is it is taken
#per location we need to do
def remove_pps_outliers(df):
    df_out=pd.DataFrame() #sub dataframe
    for key,subdf in df.groupby('location'):
        m=np.mean(subdf.price_per_sqft)
        st=np.std(subdf.price_per_sqft)
        reduced_df=subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft<= (m+st))]
        df_out=pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
# 1- variance to 1+ variance

df7=remove_pps_outliers(df6)
df7.shape
```

```python
#now we see in data that the sqft is nearby to 2bhk and 3 bhk and 2bhk have more price than 3bhk
#we need to visualize it
def plot_scatter_chart(df,location):
    bhk2 =df[(df.location==location) & (df.bhk==2)]
    bhk3 =df[(df.location==location) & (df.bhk==3)]
    matplotlib.rcParams['figure.figsize']=(15,10)
    plt.scatter(bhk2.total_sqft,bhk2.price_per_sqft,color='blue',label='2 BHK',s=50) #s --size of grapg seeing
    plt.scatter(bhk3.total_sqft,bhk3.price_per_sqft,color='green',label='3 BHK',marker='+',s=50)
    plt.xlabel("total sqft area")
    plt.ylabel(location)
    plt.title(location)
    plt.legend() # INDEX type of indicate which is what
    
plot_scatter_chart(df7,"Rajaji Nagar")
```

```python
#see at 1700 2bhk price more than 3 bhk

def plot_scatter_chart(df,location):
    bhk2 =df[(df.location==location) & (df.bhk==2)]
    bhk3 =df[(df.location==location) & (df.bhk==3)]
    matplotlib.rcParams['figure.figsize']=(15,10)
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHK',s=50) #s --size of grapg seeing
    plt.scatter(bhk3.total_sqft,bhk3.price,color='green',label='3 BHK',marker='+',s=50)
    plt.xlabel("total sqft area")
    plt.ylabel(location)
    plt.title(location)
    plt.legend() # INDEX type of indicate which is what
    
plot_scatter_chart(df7,"Hebbal")
```

```python
def remove_bhk_outliers(df):
    exclude_indices=np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats={}
        for bhk,bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean':np.mean(bhk_df.price_per_sqft),
                'std':np.std(bhk_df.price_per_sqft),
                'count':bhk_df.shape[0]
            }
        for bhk,bhk_df in location_df.groupby('bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices=np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')

df8=remove_bhk_outliers(df7)
df8.shape
            
            
```
